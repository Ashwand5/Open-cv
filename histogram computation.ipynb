```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img= cv.imread('cat_image.jpg')
cv.imshow('cat',img)
cv.waitKey(0)
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGRA2GRAY)
cv.imshow('gray',gray)
cv.waitKey(0)
```

```python
img_hist = cv.calcHist([gray],[0],None,[256],[0,256])
plt.figure()
plt.title("Image Histogram")
plt.xlabel('Bins')
plt.ylabel("No of pixels")
plt.plot(img_hist)
plt.xlim([0,256])
```

## histogram for BGR image

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load the image
img = cv.imread('cat_image.jpg')
cv.imshow('cat', img)
cv.waitKey(0)

# Calculate histograms for each color channel
colors = ('b', 'g', 'r')
for i, color in enumerate(colors):
    img_hist = cv.calcHist([img], [i], , [256], [0, 256])
    plt.plot(img_hist, color=color)
    plt.xlim([0, 256])

# Set plot title and labels
plt.title("Color Image Histogram")
plt.xlabel('Bins')
plt.ylabel('Number of Pixels')
plt.show()
```

```python
import numpy as np
```

```python
blank = np.zeros(img.shape[:2],dtype='uint8')
cv.imshow('blank', blank)
cv.waitKey(0)
```

```python
mask = cv.circle(blank,(img.shape[1]//2,img.shape[0]//2),100,255,-1)
cv.imshow('mask', mask)
cv.waitKey(0)
```

```python
masked= cv.bitwise_and(img,img,mask=mask)
cv.imshow('blank',masked)
cv.waitKey(0)
```

```python
img_hist = cv.calcHist([gray],[0],masked,[256],[0,256])
plt.figure()
plt.title("Image Histogram")
plt.xlabel('Bins')
plt.ylabel("No of pixels")
plt.plot(img_hist)
plt.xlim([0,256])
```
